A partir de los datos aportados en Kaggle, sabemos las unidades que se utilizan en cada una de las columnas, en este caso nos centraremos en utilizar el 'Air Time' que está en minutos y representa el tiempo que transcurre el avión en el aire, no nos interesa el tiempo que transcurre desde el despegue y el aterrizaje ya que no es representativo para un cálculo de velocidades.
Por otro lado, también utilizaremmos la distancia del recorrido ('Distance') que está con unidades de millas.


```python
df1['Velocidad media']=(df1['Distance']/df1['AirTime'])
df1

#La unidad utilizada para velocidad media es millas/hora
```

Para saber si los vuelos han llegado a tiempo o lo han hecho en hora, tenemos en cuenta que en la información aportada en Kaggle nos dice que los vuelos que se consideran tarde son los que tienen una llegada 15 minutos por sobre la hora de llegada estipulada por la aerolinea.


```python
df1.loc[df1['ArrDelay'] <= 15, 'Estado'] = 0 #1 equivale a 'en hora'
df1.loc[df1['ArrDelay'] > 15, 'Estado'] = 1 #0 equivale a 'tarde'
df1
```

**Fes una taula de les aerolínies amb més endarreriments acumulats.TOP5** 

```python
retrasos=df1[df1['Estado']==1].groupby(['UniqueCarrier', 'Estado'])['Estado'].count().sort_values(ascending=False).head(5)
retrasos
#Para seleccionar únicamente de Estado los que van con retraso[df1['Estado']==1]
#debo recordar que para hacer groupby es importante saber que el orden en que se ponga es que usará apra luego hacer la tabala
#como si fuese un x e y

#Cuando se utiliza el sort_values el valor del ascending es predeterminado como True, o sea ascendete, por eso
#se modifica. Y con el head seleccionamos que cantidad de datos extraer
```

Los 5 vuelos más largos en minutos son:

```python
vuelo_largo=df1.groupby(['FlightNum', 'UniqueCarrier'])['AirTime'].max().sort_values(ascending=False).head(5)
vuelo_largo
```
